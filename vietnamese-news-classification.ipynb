```python
# Visualize top feature
def plot_feature_distribution(df, column, top_n=20, figsize=(12, 8)):
    # T√≠nh to√°n s·ªë l∆∞·ª£ng m·∫´u t·ªïng c·ªông v√† top_n th·ªÉ lo·∫°i
    top_categories = df[column].value_counts().nlargest(top_n)
    total = df[column].shape[0]

    # V·∫Ω bi·ªÉu ƒë·ªì
    plt.figure(figsize=figsize)
    bars = plt.barh(top_categories.index, top_categories.values, color='skyblue')

    plt.title(f'Top {top_n} {column} Distribution')
    plt.xlabel('S·ªë l∆∞·ª£ng')
    plt.ylabel('Category')
    plt.gca().invert_yaxis()
    plt.tight_layout()

    # V·∫Ω t·ª∑ l·ªá ph·∫ßn trƒÉm tr√™n t·ª´ng c·ªôt
    for bar in bars:
        width = bar.get_width()
        percent = (width / total) * 100
        plt.text(width + 100,  # ƒêi·ªÅu ch·ªânh kho·∫£ng c√°ch text
                 bar.get_y() + bar.get_height() / 2,
                 f'{percent:.3f}%',
                 va='center')

    # Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
    plt.show()
```

```python
def plot_missing_data(df, figsize=(8, 3)):
  report = df.isna().sum()
  x = report.index
  y = report.values
  plt.figure(figsize=figsize)
  plt.title("D·ªØ li·ªáu thi·∫øu theo t·ª´ng thu·ªôc t√≠nh")
  plt.bar(x, y)
  plt.show()
```

```python
def visualize_boxplot(feature_col, group_col, df, figsize=(10, 5)):
  plt.figure(figsize=figsize)
  sns.boxplot(x=group_col, y=feature_col, data=df)
  plt.title(f'Box plot: Ph√¢n b·ªë {feature_col} theo {group_col}')
  plt.xticks(rotation=-45)
  plt.xlabel(group_col)
  plt.ylabel(feature_col)
  plt.grid(True, linestyle='--', linewidth=0.5)
  plt.tight_layout()
  plt.show()
```

```python
def build_model(model, X_train, y_train, X_test, y_test, average='macro', report=False, save_to = None):

  start_time = time.time()
  model.fit(X_train, y_train)
  end_time = time.time()

  y_pred = model.predict(X_test)
  accuracy = round(accuracy_score(y_test, y_pred), 3)
  precision = round(precision_score(y_test, y_pred, average=average), 3)
  recall = round(recall_score(y_test, y_pred, average=average), 3)
  f1 = round(f1_score(y_test, y_pred, average=average), 3)

  if(report): # Option ƒë·ªÉ hi·ªÉn th·ªã th√¥ng tin c√°c metrics c·ªßa model
    training_time = end_time - start_time
    print(f"Training time: {str(timedelta(seconds=int(training_time)))}")
    print(classification_report(y_test, y_pred))

  if(save_to != None): # Option ƒë·ªÉ save l·∫°i model sau khi train
    saved_location = joblib.dump(model, save_to)
    print('Model saved at:', saved_location)

  return model, (accuracy, precision, recall, f1)
```

```python
def build_model_with_data_generator(*,
  model=None,
  batch_size=None,
  steps_per_epoch=None,
  validation_steps=None,
  train_generator=None,
  val_generator=None,
  monitor='val_loss',
  save_to='best_model.keras',
  epochs=10,
  verbose=1,
  patience=3,
):

  # callbacks
  early_stop = EarlyStopping(
    monitor=monitor,
    patience=patience,
    restore_best_weights=True,
    verbose=verbose
  )

  checkpoint = ModelCheckpoint(
    filepath=os.path.join(save_to),
    monitor=monitor,
    save_best_only=True,
    save_weights_only=False,
    verbose=verbose
  )

  # train
  history = model.fit(
      train_generator,
      steps_per_epoch=steps_per_epoch,
      epochs=epochs,
      verbose=verbose,
      validation_data=val_generator,
      validation_steps=validation_steps,
      callbacks=[early_stop, checkpoint],
  )

  return history
```

```python
with open(os.path.join(WORKING_DRIVE, 'vietnamese-stopwords-dash.txt'), 'r', encoding='utf-8') as f:
    stopwords = set(line.strip() for line in f if line.strip())

def remove_stopwords(text, stopwords):
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token not in stopwords]
    return ' '.join(filtered_tokens)
```

```python
def filter_noun_verb(txt):
  pos_tagged = pos_tag(txt)
  words = [word for word, pos in pos_tagged if pos in ["N", "V"]]
  return " ".join(words)
```

```python
# Danh s√°ch ch·ªØ s·ªë ·ªü d·∫°ng ch·ªØ
number_words = {"m·ªôt", "hai", "ba", "b·ªën", "nƒÉm", "s√°u", "b·∫£y", "t√°m", "ch√≠n", "m∆∞·ªùi", "m∆∞∆°i", "m·ªët", "trƒÉm", "ngh√¨n", "tri·ªáu", "t·ª∑"}
# pattern c√°c ch·ªØ c√°i c·∫ßn gi·ªØ l·∫°i
pattern = r"[^a-zA-Z√°√†·∫£√£·∫°ƒÉ√¢·∫Ø·∫±·∫≥·∫µ·∫∑·∫•·∫ß·∫©·∫´·∫≠√©√®·∫ª·∫Ω·∫π√™·∫ø·ªÅ·ªÉ·ªÖ·ªá√≠√¨·ªâƒ©·ªã√≥√≤·ªè√µ·ªç√¥·ªë·ªì·ªï·ªó·ªô∆°·ªõ·ªù·ªü·ª°·ª£√∫√π·ªß≈©·ª•∆∞·ª©·ª´·ª≠·ªØ·ª±√Ω·ª≥·ª∑·ªπ·ªµƒë_\s]"

def text_preprocess(document):
  # chu·∫©n h√≥a unicode
  document = covert_unicode(document)
  # t√°ch t·ª´
  document = word_tokenize(document, format="text")
  # ƒë∆∞a v·ªÅ lowercase
  document = document.lower()
  # lo·∫°i b·ªè k√Ω t·ª± ƒë·∫∑c bi·ªát v√† s·ªë
  document = re.sub(pattern, " ", document)
  # lo·∫°i b·ªè c√°c ch·ªØ s·ªë ·ªü d·∫°ng ch·ªØ
  document = " ".join([word for word in document.split() if word not in number_words])
  # lo·∫°i b·ªè stopwords
  document = remove_stopwords(document, stopwords)
  # l·ªçc ra danh t·ª´ v√† ƒë·ªông t·ª´
  document = filter_noun_verb(document)
  # lo·∫°i b·ªè kho·∫£ng tr·∫Øng d∆∞ th·ª´a
  document = re.sub(r'\s+', ' ', document).strip()
  return document
```

```python
# S·ª≠ d·ª•ng callback t·ª± ƒë·ªông l∆∞u l√™n Drive
class DriveSaveCallback(TrainerCallback):
    def on_save(self, args, state, control, **kwargs):
        # Sao ch√©p checkpoint m·ªõi nh·∫•t l√™n Drive
        checkpoint_dir = f"checkpoint-{state.global_step}"
        source = os.path.join(args.output_dir, checkpoint_dir)
        destination = os.path.join(CLASSIFIER_DIR, "drive_backups", checkpoint_dir)

        if os.path.exists(destination):
            shutil.rmtree(destination)

        shutil.copytree(source, destination)
        print(f"ƒê√£ sao l∆∞u checkpoint {state.global_step} l√™n Drive")

# Callback d·ª´ng s·ªõm khi val_loss kh√¥ng c·∫£i thi·ªán sau N steps
early_stopping = EarlyStoppingCallback(
    early_stopping_patience=5,
    early_stopping_threshold=0.001
)

# T·∫°o Hugging Face Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[early_stopping, DriveSaveCallback]
)
```

### Data generator

```python
def embedding_generator(X, y, w2v, batch_size=128, max_len=300):
    embedding_dim = w2v.vector_size
    X = X.to_list()
    num_samples = len(X)
    indices = np.arange(num_samples)

    while True:
        np.random.shuffle(indices)  # shuffle m·ªói epoch

        for start in range(0, num_samples, batch_size):
            end = min(start + batch_size, num_samples)
            batch_indices = indices[start:end]
            current_batch_size = end - start

            X_batch = np.zeros((batch_size, max_len, embedding_dim), dtype=np.float32)
            y_batch = np.zeros((batch_size, y.shape[1]), dtype=np.float32)

            for b, idx in enumerate(batch_indices):
                words = X[idx].split()
                for j, word in enumerate(words[:max_len]):
                    if word in w2v.wv:
                        X_batch[b, j] = w2v.wv[word]
                y_batch[b] = y[idx]

            yield X_batch[:current_batch_size], y_batch[:current_batch_size]
```

```python
def predict_generator(X, w2v, batch_size=128, max_len=300):
    embedding_dim = w2v.vector_size
    X = X.to_list()
    num_samples = len(X)
    index = 0

    while index < num_samples:
        X_batch = np.zeros((batch_size, max_len, embedding_dim), dtype=np.float32)

        actual_batch_size = min(batch_size, num_samples - index)

        for b in range(actual_batch_size):
            words = X[index].split()
            for j, word in enumerate(words[:max_len]):
                if word in w2v.wv:
                    X_batch[b, j] = w2v.wv[word]
            index += 1

        yield (X_batch[:actual_batch_size],)
```

### BiLSTM (Average Pooling)

```python
max_len = 300
embedding_dim = 100

input = Input(shape=(max_len, embedding_dim)) #(batch, 300, 100)                    # params = 0

                                                                                    # params = 4 gates * [(hidden units + inputs) * hidden units + bias] * 2
lstm = Bidirectional(LSTM(128, return_sequences=True))(input) #(batch, 300, 256)    # 234496 =    4    * [(     128     +   100 ) *     128      +  128] * 2

pooling = GlobalAveragePooling1D()(lstm) #(batch, 256)                              # params = 0

dropout = Dropout(0,3)(pooling) #(batch, 256)                                       # params = 0
                                                                                    # params = inputs * current neurons + bias
output = Dense(9, activation='softmax')(dropout) #(batch, 9)                        #  2313  =  256   *         9       +   9

model = Model(inputs=input, outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Visualize top feature
def plot_feature_distribution(df, column, top_n=20, figsize=(12, 8)):
    # T√≠nh to√°n s·ªë l∆∞·ª£ng m·∫´u t·ªïng c·ªông v√† top_n th·ªÉ lo·∫°i
    top_categories = df[column].value_counts().nlargest(top_n)
    total = df[column].shape[0]

    # V·∫Ω bi·ªÉu ƒë·ªì
    plt.figure(figsize=figsize)
    bars = plt.barh(top_categories.index, top_categories.values, color='skyblue')

    plt.title(f'Top {top_n} {column} Distribution')
    plt.xlabel('S·ªë l∆∞·ª£ng')
    plt.ylabel('Category')
    plt.gca().invert_yaxis()
    plt.tight_layout()

    # V·∫Ω t·ª∑ l·ªá ph·∫ßn trƒÉm tr√™n t·ª´ng c·ªôt
    for bar in bars:
        width = bar.get_width()
        percent = (width / total) * 100
        plt.text(width + 100,  # ƒêi·ªÅu ch·ªânh kho·∫£ng c√°ch text
                 bar.get_y() + bar.get_height() / 2,
                 f'{percent:.1f}%',
                 va='center')

    # Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
    plt.show()
```

```python
with open(vietnamese_stopwords_path, 'r', encoding='utf-8') as f:
    stopwords = set(line.strip() for line in f if line.strip())

def remove_stopwords(text, stopwords):
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token not in stopwords]
    return ' '.join(filtered_tokens)
```

```python
def filter_noun_verb(txt):
  pos_tagged = pos_tag(txt)
  words = [word for word, pos in pos_tagged if pos in ["N", "V"]]
  return " ".join(words)
```

```python
# Danh s√°ch ch·ªØ s·ªë ·ªü d·∫°ng ch·ªØ
number_words = {"m·ªôt", "hai", "ba", "b·ªën", "nƒÉm", "s√°u", "b·∫£y", "t√°m", "ch√≠n", "m∆∞·ªùi", "m∆∞∆°i", "m·ªët", "trƒÉm", "ngh√¨n", "tri·ªáu", "t·ª∑"}
# pattern c√°c ch·ªØ c√°i c·∫ßn gi·ªØ l·∫°i
pattern = r"[^a-zA-Z√°√†·∫£√£·∫°ƒÉ√¢·∫Ø·∫±·∫≥·∫µ·∫∑·∫•·∫ß·∫©·∫´·∫≠√©√®·∫ª·∫Ω·∫π√™·∫ø·ªÅ·ªÉ·ªÖ·ªá√≠√¨·ªâƒ©·ªã√≥√≤·ªè√µ·ªç√¥·ªë·ªì·ªï·ªó·ªô∆°·ªõ·ªù·ªü·ª°·ª£√∫√π·ªß≈©·ª•∆∞·ª©·ª´·ª≠·ªØ·ª±√Ω·ª≥·ª∑·ªπ·ªµƒë_\s]"

def text_preprocess(document):
  # chu·∫©n h√≥a unicode
  document = covert_unicode(document)
  # t√°ch t·ª´
  document = word_tokenize(document, format="text")
  # ƒë∆∞a v·ªÅ lowercase
  document = document.lower()
  # lo·∫°i b·ªè k√Ω t·ª± ƒë·∫∑c bi·ªát v√† s·ªë
  document = re.sub(pattern, " ", document)
  # lo·∫°i b·ªè c√°c ch·ªØ s·ªë ·ªü d·∫°ng ch·ªØ
  document = " ".join([word for word in document.split() if word not in number_words])
  # lo·∫°i b·ªè stopwords
  document = remove_stopwords(document, stopwords)
  # l·ªçc ra danh t·ª´ v√† ƒë·ªông t·ª´
  document = filter_noun_verb(document)
  # lo·∫°i b·ªè kho·∫£ng tr·∫Øng d∆∞ th·ª´a
  document = re.sub(r'\s+', ' ', document).strip()
  return document
```

```python
tf_idf_sums = X_tf_idf.sum(axis=0).A1
vocab = vectorizer.get_feature_names_out()

# Gh√©p t·ª´ v·ªõi t·ªïng TF-IDF
tf_idf_scores = list(zip(vocab, tf_idf_sums))
tf_idf_sorted = sorted(tf_idf_scores, key=lambda x: x[1], reverse=True)

# Hi·ªÉn th·ªã top 20 t·ª´ c√≥ t·ªïng TF-IDF cao nh·∫•t
print(pd.DataFrame(tf_idf_sorted[:20], columns=["T·ª´", "T·ªïng TF-IDF"]))
```
